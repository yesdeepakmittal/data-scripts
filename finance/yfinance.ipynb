```python
import datetime
import pandas as pd

stocks = ['ADBE','FB','RELIANCE.NS','GOOG','SBIN.NS','9988.HK','BRK-A']
start = datetime.datetime.today() - datetime.timedelta(360)
end = datetime.datetime.today()
df = pd.DataFrame()

for stock in stocks:
    df[stock] = yf.download(stock,start=start,end=end)['Adj Close']
df.tail()
```
