```python
import pandas as pd
from yahoofinancials import YahooFinancials
import datetime
```

```python
ticker= 'ADBE'
yahoo_financials = YahooFinancials(ticker)
data = yahoo_financials.get_historical_price_data('2021-04-01','2021-04-30','daily')
data
```

```python
stocks = ['ADBE','FB','RELIANCE.NS','GOOG','SBIN.NS','9988.HK','BRK-A']

df = pd.DataFrame()
end = (datetime.date.today()).strftime('%Y-%m-%d')
start = (datetime.date.today()-datetime.timedelta(1825)).strftime('%Y-%m-%d')
for stock in stocks:
    yahoo_financials = YahooFinancials(stock)
    obj = yahoo_financials.get_historical_price_data(start,end,'daily')
    ohlv = obj[stock]['prices']
    temp = pd.DataFrame(ohlv)[['formatted_date','adjclose']]
    temp.set_index('formatted_date',inplace=True)
    temp.dropna(inplace=True)
    df[stock] = temp['adjclose']
```
